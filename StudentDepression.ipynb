```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("studentDepression.csv")
df=df.drop(columns=["id"])
df.head(5)
```

```python
print(df["Dietary Habits"].unique())
print(df["Degree"].unique())
print(df["City"].unique())
print(df["Sleep Duration"].unique())
```

```python
ordinalEncoder=OrdinalEncoder()
labelEncoder=LabelEncoder()
df[["Dietary Habits"]]=ordinalEncoder.fit_transform(df[["Dietary Habits"]])
df[["Sleep Duration"]]=ordinalEncoder.fit_transform(df[["Sleep Duration"]])
df["Gender"]=df["Gender"].replace({"Male":1,"Female":0})
df=df.rename(columns={"Have you ever had suicidal thoughts ?":"Suicide","Family History of Mental Illness":"Family History"})
df["Suicide"]=df["Suicide"].replace({"Yes":1,"No":0})
df["Family History"]=df["Family History"].replace({"Yes":1,"No":0})
df["City"]=labelEncoder.fit_transform(df["City"])
df["Degree"]=labelEncoder.fit_transform(df["Degree"])
df["Profession"]=labelEncoder.fit_transform(df["Profession"])
df.head(5)
```

```python
rfc=RandomForestClassifier()
x=df.drop(columns=["Depression"])
y=df["Depression"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
rfc.fit(x_train,y_train)
rfc.score(x_test,y_test)
#rfc.predict([[1,19,1,11,3.0,0.0,7.7,3.0,0.0,1,1,3,1,3.0,3.0,1,3]]) mine data :/
```
